```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

This markdown represents a Python code snippet defining a `PromptTemplate` object. The template string inside the `PromptTemplate` is structured to guide the recommender system in analyzing and rating a given review.

```python
prompt = PromptTemplate(
    input_variables=['review'],
    template = '''You have to work as a recommender system on a real time basis {review} You have to read the review {review} and analyse it
    and rate it according to your intelligence into 
    Good if the review is good represent it by 1
    Neutral if the review is good represent it by 0
    Bad if the review is good represent it by -1
    and based on it generate the analysis
    and just give me the numeric rating i.e., 0 for neutral 
    1 for good
    and -1 for bad
    '''
)
```

## This is the prompt template

```python
chain = LLMChain(llm=llm,prompt=prompt)
```

```python
result = chain.run('Fantastic presentation and service.').strip()
```
